# Interencia

Buscaremos modelar un experimento de doble rendija y que muestre el patron generado a partir de los par치metros puestos

```python
import numpy as np
import matplotlib.pyplot as plt

def double_slit(screen_distance, slit_distance, wavelength, slit_width, screen_width, num_points):
    # Constantes
    k = 2 * np.pi / wavelength  # Numero de onda
    d = sd  # Distancia entre rendijas
    a = sld  # Ancho de cada rendija
    L = sd  # Distancia a la pantalla
    w = scw  # Ancho de la pantalla

    x_val = np.linspace(-w / 2, w / 2, num_points)

    # Intensidad
    intensidad = np.zeros_like(x_val, dtype=float)

    for x_index, x in enumerate(x_val):
        # Calculate the path difference for each point on the screen
        path_difference = d * np.sin(np.arctan(x / L))

        # Calculate the amplitude for each point on the screen
        amplitud = np.cos((k * a / 2) * np.sin(path_difference))  # Amplitude for a single slit

        # Calculate intensity (intensity is proportional to the square of amplitude)
        intensidad[x_index] = amplitud**2

    return (x_val, intensidad)

def plot_double_slit_interference(x_val, intensity):
    plt.plot(x_val, intensity, label='Intensidad')
    plt.title('Patr칩n de interferencia')
    plt.xlabel('Posici칩n')
    plt.ylabel('Intensidad')
    plt.legend()
    plt.show()

# Parametros
sd = 0.55  # Distancia de la rendija a la pantalla
sld = 0.85  #Distancia entre rendijas
wavelength = 0.0810  # Longitud de onda de la luz
sw = 0.25  # Ancho de la rendija
scw = 1  # Ancho de la pantalla
num_points = 20000  # Number of points on the screen

# Mostrar el patr칩n de interferencia
x_values, intensity = double_slit(sd, sld, wavelength, sw, scw, num_points)
plot_double_slit_interference(x_values, intensity) # Changed x_val to x_values
```
